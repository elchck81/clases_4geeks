```python
#algoritmo de suma y multiplicación de matrices
matriz_a = [[-1, 0, 2],[0, 1, 2],[0, 0, -1]]

matriz_a[0][0]
```

```python
def imprimir_matriz(matriz:list)->None:
  for e in range(len(matriz)):
    for j in range(len(matriz[0])):
      print(matriz[e][j], end=" ")
    print("\n")

imprimir_matriz(matriz_a)
```

Dada una lista simple mi_lista = [3, 4,7, -1], genere un algoritmo para imprimir cada elemento de la lista separado por dos simbolos de igual (==)

```python
mi_lista = [3, 4, 7, -1]

def leer_lista(matriz:list)->None:
  for i in range(len(matriz)-1):
    print(matriz[i], end=" == ")
  print(matriz[-1])

leer_lista(mi_lista)
```

Dados dos matrices, genere un algoritmo para sumar los elementos de la matriz:  m_1 = [[0,1],[3,-2]]  
m_2 = [[0,-1],[1,1]]

```python
m_1 = [[0,1],[3,-2]]
m_2 = [[0,-1],[1,1]]

def suma_matrices(matriz1:list, matriz2:list)->list:
  suma_matriz = matriz1.copy()
  for i in range(len(matriz1)):
    for j in range(len(matriz1)):
       suma_matriz[i][j] = matriz1[i][j] + matriz2[i][j]
  return suma_matriz

suma_matrices(m_1,m_2)

```

Dada una lista de numeros, retornar los numeros elevados al cuadrado. Ej. lista_b = [-3, 2, 0, 10]

```python
lista_b = [-3, 2, 0, 10]


def lista_cuadrado(lista:list)->list:
  lista_result = []
  for i in lista:
      lista_result.append(i*i)
  return lista_result
lista_cuadrado(lista_b)
```

```python
[x*x for x in lista_b]
```

```python
[x-1 for x in lista_b]
```

Elevar numeros al cuadrado solo para numeros mayores a 0

```python
[x*x if x > 0 else x for x in lista_b ]
```
