Ha quedado bien establecido por los cientificos que si el contenido de nicotina de los cigarrillos es de 25mg o más, es posible que el fumador contraiga cáncer. Si el contenido de nicotina medio es menor de 25mg, el usuario se encuentra relativamente seguro.

Para establecer la hipótesis nuestro interés es muestrear una cantidad de cigarrillos y ver si el promedio de esta muestra es mayor a 25.

H0: $\mu$ >= 25  
H1: $\mu$ < 25

Un proceso industrial usado por una fábrica durante algunos años da una producción media de 100 unidades por hora, con una desviación estándar de 8 unidades. Acaba de ponerse en el mercado una nueva máquina para producir el mismo producto. Aunque es muy costosa comparada con la actual, su adopción sería muy lucrativa si su producción media fuera mayor a 150 unidades por hora. Suponga que la nueva máquina ha sido probada 35 horas y se ha encontrado que el promedio de producción es de 160 unidades por hora. Explicar su decisión respecto a la compra de la nueva máquina.

H0 = $\mu$ <= 150  
H1 = $\mu$ > 150  
$\bar{x}$ = 160  
$σ$ = 8  
n = 35

z = $\frac{\bar{x} - \mu0}{σ/\sqrt{n}}$  



```python
n = 35
sample_mean = 160
poblational_mean = 150
poblational_std = 8
alfa = 0.05 #esto significa estamos dispuestos a aceptar un error de 5%

z_score = (sample_mean - poblational_mean) / (poblational_std / np.sqrt(n))
z_score #cantidad de desviación estándar alejado de la media

p_value = 1 - stats.norm.cdf(abs(z_score))
p_value

#Si el p_value es menor que alfa entonces queda H1 sino queda H0
#Nos quedamos con H1 Ya que p_value es menor a alfa
```

La distancia media que debe recorrer un automovil de cierta marcapara detenerse cuando marcha a 30 mph es 65 pies. El departamento de ingenieria de la compañia ha diseñado un nuevo sistema de frenos considerado como más eficaz. Para probar esta invención es instalado en 64 autos; las pruebas demuestran que la distancia media para detener un automovil a una velocidad media de 30 millas por hora es 63.5 pies, con una desviación estándar de 4 pies. Esta reducción de distancia ¿Demuestra que el nuevo sistema de frenos es más eficaz que el antiguo?

H0:$\mu$ >= 65  
H1:$\mu$ < 65  
Nivel de significancia: $\alpha$ = 0.01

```python
n = 64
sample_mean = 63.5
poblational_mean = 65
poblational_std = 4
alfa = 0.01 #esto significa estamos dispuestos a aceptar un error de 1%

z_score = (sample_mean - poblational_mean) / (poblational_std / np.sqrt(n))
z_score #cantidad de desviación estándar alejado de la media

p_value = 1 - stats.norm.cdf(abs(z_score))
display(p_value)
```

```python
n = 100
sample_mean = 585
poblational_mean = 600
poblational_std = 80
alfa = 0.05 #esto significa estamos dispuestos a aceptar un error de 1%

z_score = (sample_mean - poblational_mean) / (poblational_std / np.sqrt(n))
z_score #cantidad de desviación estándar alejado de la media

p_value = 1 - stats.norm.cdf(abs(z_score))
display(p_value)
```

A farmer decides to test three different types of fertilizers to determine if one is superior in terms of corn production. The farmer plants corn on 15 identical plots and uses all three fertilizers (5 plots for each type). At the end of the season, he measures the corn yield (in kg) of each plot, with the following result:

|Fertilizer 1|	Fertilizer 2|	Fertilizer 3|
|-|-|-|
|20|	22|	24|
|21|	21|	23|
|20|	23|	22|
|19|	22|	23|
|20|	21|	24|

With this data, he seeks to answer the following question: Is there a significant difference in average corn yield between the three types of fertilizers?

To help you, follow the points below:

State the hypothesis: null and alternative hypothesis.  
Perform the ANOVA test.  
Analyze the conclusions.  
If one fertilizer is better than another, how can we know it?

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
data = [20,21,20,19,20,22,21,23,22,21,24,23,22,23,24]
labels = ["f1","f1","f1","f1","f1","f2","f2","f2","f2","f2","f3","f3","f3","f3","f3"]
result = pairwise_tukeyhsd(data, labels, alpha=0.05)
print(result)
```

Dos grupos de personas han seguido dos dietas distintas y con ellas bajaron de peso, el primer grupo siguió la dieta 1 la cual tiene un promedio de pérdida de peso de 2.55 kg por mes y el segundo de 3.0 kg por mes. Hay una diferencia significativa en el promedio de pérdida de peso entre el grupo que siguió la dieta 1 y el grupo que siguió la dieta 2?

```python
import pandas as pd
import scipy.stats as stats
import numpy as np

diet1 = [2.0, 2.5, 3.0, 2.8, 2.3, 2.7, 2.5]
diet2 = [3.0, 3.2, 3.1, 2.9, 2.8, 3.0, 3.2]
d1 = pd.DataFrame(diet1)
d2 = pd.DataFrame(diet2)

#H0: No hay una diferencia significativa
#H1: Hay una diferencia significativa

n = 7
sample_mean = d2.mean()
poblational_mean = d1.mean()
poblational_std = d1.std()
alfa = 0.05

z_score = (sample_mean - poblational_mean) / (poblational_std / np.sqrt(n))
z_score #cantidad de desviación estándar alejado de la media

p_value = 1 - stats.norm.cdf(abs(z_score))
display(p_value)
```
