- 1. cual es la probabilidad de obtener un numero impar en un lanzamiento de dados

```python
s = {1,2,3,4,5,6}
e = {1,3,5}

len(set.intersection(e,s))
Fraction(len(set.intersection(e,s)),len(s))
```

- 2. Cual es la probabilidad de obtener un numero primo o un numero par en un lanzamiento de dados

```python
s = {1,2,3,4,5,6}
e1 = {2,3,5}
e2 = {2,4,6}
e = set.union(e1,e2)

Fraction(len(set.intersection(e,s)),len(s))
```

- 3 cual es la probabilidad de obtener un numero par y primo en un lanzamiento de dados

```python
s = {1,2,3,4,5,6}
e1 = {2,3,5}
e2 = {2,4,6}
e = set.intersection(e1,e2)

Fraction(len(set.intersection(e,s)),len(s))
```

```python
def pr(event:set,space:set)->Fraction:
    return Fraction(len(set.intersection(event,space)),len(space))
```

```python
pr({2,4,6},{1,2,3,4,5,6})
```

El metodo de montecarlo nos permite aproximar una probabilidad en base a la repeticion continua de un evento

- Cual es la probabilidad de obtener un numero par en el lanzamiento de un dado

```python
event = [2,4,6]
space = [1,2,3,4,5,6]
frec = []
trials = range(1,1001)
for trial in trials:
    total = 0
    for i in range(trial):
        if random.choice(space) in event:
            total = total + 1
    frec.append(total/trial)
```

```python
plt.plot(trials,frec)
plt.axhline(y=0.5,color="tomato")
```
