```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, poisson
```

El modelo Bernoulli se aplica a un random variable que puede asumir sólo dos valores. Para este modelo tenemos:

- $E(X)=p$ 
- $V(X)=pq$

Donde p es la probabilidad de éxito y q es la probabilidad de fracaso. $E(X)$ significa la esperanza de la variable aleatoria X y $V(X)$ significa la varianza de la variable aleatoria X

El modelo binomial puede considerarse como la suma de n variables Bernoulli independientes una de otra donde se cumpla lo siguiente:
1. Debe haber un número fijo de pruebas repetidas estadísticamente independientes.
2. Cada prueba debe ser una variable Bernoulli, es decir puede resultar en un éxito o en un fracaso.
3. Todas las pruebas deben tener identicas probabilidades de éxito.

El modelo binomial es discreto y tiene lo siguiente:
- $E(X)=np$
- $V(X)=npq$

Donde n es el número de ensayos o pruebas, p es la probabilidad de éxito y q es la probabilidad de fracaso.

Una moneda es lanzada 10 veces. ¿Cuál es la probabilidad de obtener 8 o más caras?

```python
n = 10
p = 0.5
# P(X >= 8) = P(X = 8) + P(X = 9) + P(X = 10) = 1 - P(X <= 7) = 1 - cdf(7)
prob = 1 - binom.cdf(6, n, p) # cdf = es la función de distribución acumulada que indica P(X = 0) + P(X = 1) + ... + P(X = 7)
prob
```

Se sabe que 90% de los trabajadores de una ciudad industrial son miembros de un sindicato. En un estudio de situación laboral de la ciudad un economista escoge al azar una muestra de 15 obreros para entrevistas. ¿Cuál es la probabilidad de que la muestra contenga 12 o menos miembros del sindicato?

```python
p = 0.9
n = 15 
# P(X <= 12) = P(X = 0) + P(X = 1) + ... + P(X = 12) = cdf(12)
prob = binom.cdf(12, n, p)
prob
```

Muchos experimentos no ocurren como un resultado definido de pruebas sino como puntos de tiempo, espacio o volumen al azar. El modelo Poisson es adecuado en estos casos si:
1. El número de ocurrencias del hecho es independiente de una unidad
2. El valor esperado de la variable es proporcional al tamaño de la unidad especificada
3. La probabilidad de mas de una ocurrencia del hecho en una unidad especificada muy pequeña es despreciable

Para este modelo tenemos:
- $E(X)=\lambda$
- $V(X)=\lambda$

Donde $\lambda$ es igual a el número de ocurrencias por unidad especificada. La distribución Poisson es discreta

En un callcenter ocurren 4 llamadas por minuto ($\lambda$)=4. ¿Cuál es la probabilidad de que no ocurran más de 2 llamadas durante los 2 minutos siguientes?

```python
n = 2 #numero de éxitos
plambda = 8 #ocurrencia en una unidad especificada 4 llamadas por minuto (2 minutos * 4 llamadas) = 8 llamadas en 2 minutos
#P(X<=2) es decir P(X=0) + P(X=1) + P(X=2)
prob = poisson.cdf(n, plambda)
prob
```

```python
#La P(X>=4) = P(X=4) + P(X=5) + P(X=6) ... P(X=inf)
#P(X>=4) = 1 - P(X<=3) = 1 - P(X=0) -  P(X=1) - P(X=2) - P(X=3)
n = 3
plambda = 8
prob = 1-poisson.cdf(n, plambda)
prob
```

Un fabricante de tejidos de lana afirma que el promedio de defectos en sus productos es 1 por 2 yardas cuadradas. Se extrajo una yarda cuadrada al azar y tenía 3 defectos. ¿Cuál es la probabilidad de obtener 3 o más defectos en cualquier yarda cuadrada si la afirmación del fabricante fuera válida?

```python
plambda = 0.5 #ocurrencia en una unidad especificada 1 falla de 2 yardas
n = 3 #defectos buscados
#P(X>=3) = P(X=3) + P(X=4) + P(X=5) ... P(X=inf)
#P(X>=3) = 1 - P(X<=2) = 1 - P(X=0) - P(X=1) - P(X=2)
prob = 1-poisson.cdf(n, plambda)
prob
```

El modelo de probabilidad más frecuentemente usado en el análisis económico y comercial es la distribución normal, que puede expresarse en las formas general y estandarizada. Se dice que una variable aleatoria X tiene una distribución normal si es continua, si está en el rango del menos infinito al infinito y si puede expresarse por su media $\mu$ y su desviación estándar $\sigma$

```python
x = np.linspace(-5, 5, 1000)
pdf = norm.pdf(x)
plt.plot(x, pdf, c="tomato")
plt.title("PDF Distribución normal")
plt.xlabel("value")
plt.ylabel("density")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-5, 5, 1000)
cdf = norm.cdf(x)
plt.plot(x, cdf, c="tomato")
plt.title("CDF Distribución normal")
plt.xlabel("value")
plt.ylabel("probability")
plt.grid(True)
plt.show()
```

```python
n = 10
p = 0.5
x = range(n+1)

pmf = binom.pmf(x, n, p)
plt.stem(x, pmf)
plt.grid(True)
plt.show()
```

```python
n = 50
plambda = 5
x = range(n+1)
pmf = poisson.pmf(x, plambda)
plt.stem(x, pmf)
plt.grid(True)
plt.show()
```
