```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open("/content/Goose.png")
```

```python
img_arr = np.array(img)
img_arr
```

```python
plt.imshow(img_arr)
plt.show()
```
